```python
import os
import cv2
from ultralytics import YOLO

# -----------------------------
# USER SETTINGS
# -----------------------------
MODEL_PATH = "/kaggle/working/feature.pt"     # your YOLO model path
INPUT_DIR = "/kaggle/working/patch_dataset/test/images"          # folder with full currency images
OUTPUT_DIR = "/kaggle/working/siamese_features"    # will contain cropped features

# Create output subfolders
FEATURE_NAMES = ["watermark_window", "security_thread", "number_panel"]
for f in FEATURE_NAMES:
    os.makedirs(os.path.join(OUTPUT_DIR, f), exist_ok=True)

# -----------------------------
# LOAD YOLO MODEL
# -----------------------------
model = YOLO(MODEL_PATH)

# -----------------------------
# PROCESS IMAGES
# -----------------------------
image_list = [f for f in os.listdir(INPUT_DIR)
              if f.lower().endswith((".png", ".jpg", ".jpeg"))]

print(f"Found {len(image_list)} images.")

for img_name in image_list:
    img_path = os.path.join(INPUT_DIR, img_name)
    img = cv2.imread(img_path)

    if img is None:
        print(f"Could not read: {img_path}")
        continue

    # Run detection
    results = model(img)[0]

    # Loop through detections
    for box in results.boxes:
        cls = int(box.cls[0])
        conf = float(box.conf[0])
        label = model.names[cls]

        # Skip if unexpected class
        if label not in FEATURE_NAMES:
            continue

        # Extract bounding box
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Clip coordinates inside image
        h, w = img.shape[:2]
        x1 = max(0, x1); y1 = max(0, y1)
        x2 = min(w, x2); y2 = min(h, y2)

        # Crop the feature region
        crop = img[y1:y2, x1:x2]

        if crop.size == 0:
            continue

        # Build save path
        out_name = img_name.split('.')[0] + f"_{label}.png"
        out_path = os.path.join(OUTPUT_DIR, label, out_name)

        cv2.imwrite(out_path, crop)
        print(f"Saved: {out_path}")

print("\nFeature extraction completed successfully!")
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Input real features folder
REAL_DIR = "/kaggle/working/siamese_features"       # each subfolder = feature name
FAKE_DIR = "/kaggle/working/fake_features"

os.makedirs(FAKE_DIR, exist_ok=True)

def add_noise(img):
    noise = np.random.normal(0, 10, img.shape).astype(np.uint8)
    return cv2.add(img, noise)

def blur(img):
    return cv2.GaussianBlur(img, (7, 7), 0)

def brighten(img):
    return cv2.convertScaleAbs(img, alpha=1.5, beta=20)

def darken(img):
    return cv2.convertScaleAbs(img, alpha=0.6, beta=-30)

def distort(img):
    h, w = img.shape[:2]
    pts1 = np.float32([[0,0],[w,0],[0,h],[w,h]])
    pts2 = pts1 + np.random.randint(-10, 10, size=(4,2)).astype(np.float32)
    M = cv2.getPerspectiveTransform(pts1, pts2)
    return cv2.warpPerspective(img, M, (w,h))

# All augmentation methods
AUGS = [add_noise, blur, brighten, darken, distort]

# Loop through feature folders
for feature in os.listdir(REAL_DIR):
    real_feature_path = os.path.join(REAL_DIR, feature)
    fake_feature_path = os.path.join(FAKE_DIR, feature)
    os.makedirs(fake_feature_path, exist_ok=True)

    images = [f for f in os.listdir(real_feature_path) if f.endswith((".jpg",".png"))]

    for img_name in tqdm(images, desc=f"Generating fake for {feature}"):
        img_path = os.path.join(real_feature_path, img_name)
        img = cv2.imread(img_path)

        # Apply all augmentations
        for i, aug in enumerate(AUGS):
            fake_img = aug(img)
            out_path = os.path.join(fake_feature_path, img_name.replace(".png","") + f"_fake{i}.png")
            cv2.imwrite(out_path, fake_img)

print("Fake feature generation completed!")
```

```python
import os
import cv2
import random
import numpy as np
from tqdm import tqdm

REAL_DIR = "/kaggle/working/siamese_features"
FAKE_DIR = "/kaggle/working/fake_features"
OUT_DIR = "/kaggle/working/siamese_pairs"

os.makedirs(OUT_DIR, exist_ok=True)

pair_images_1 = []
pair_images_2 = []
pair_labels = []     # 1 = same(real-real), 0 = fake(real-fake)

IMG_SIZE = 128

def load_and_resize(path):
    img = cv2.imread(path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    return img.astype("float32") / 255.0

# Loop through features
features = os.listdir(REAL_DIR)

for feature in features:

    real_list = [os.path.join(REAL_DIR, feature, f)
                 for f in os.listdir(os.path.join(REAL_DIR, feature))]
    fake_list = [os.path.join(FAKE_DIR, feature, f)
                 for f in os.listdir(os.path.join(FAKE_DIR, feature))]

    # Positive pairs (real vs real)
    for i in tqdm(range(len(real_list)-1), desc=f"Positive pairs {feature}"):
        img1 = load_and_resize(real_list[i])
        img2 = load_and_resize(real_list[i+1])

        pair_images_1.append(img1)
        pair_images_2.append(img2)
        pair_labels.append(1)

    # Negative pairs (real vs fake)
    for i in tqdm(range(min(len(real_list), len(fake_list))), desc=f"Negative pairs {feature}"):
        img1 = load_and_resize(real_list[i])
        img2 = load_and_resize(fake_list[i])

        pair_images_1.append(img1)
        pair_images_2.append(img2)
        pair_labels.append(0)

# Save dataset
np.save(os.path.join(OUT_DIR, "pair_img1.npy"), np.array(pair_images_1))
np.save(os.path.join(OUT_DIR, "pair_img2.npy"), np.array(pair_images_2))
np.save(os.path.join(OUT_DIR, "pair_labels.npy"), np.array(pair_labels))

print("Siamese pair dataset created successfully!")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model

# Load data
x1 = np.load("/kaggle/working/siamese_pairs/pair_img1.npy")
x2 = np.load("/kaggle/working/siamese_pairs/pair_img2.npy")
y  = np.load("/kaggle/working/siamese_pairs/pair_labels.npy")

# Siamese base CNN encoder
def build_encoder():
    inp = layers.Input(shape=(128,128,3))
    x = layers.Conv2D(32, (3,3), activation="relu")(inp)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(64, (3,3), activation="relu")(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(128, (3,3), activation="relu")(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation="relu")(x)
    return Model(inp, x)

encoder = build_encoder()

# Siamese Network
input_a = layers.Input(shape=(128,128,3))
input_b = layers.Input(shape=(128,128,3))

emb_a = encoder(input_a)
emb_b = encoder(input_b)

# L1 distance
distance = layers.Lambda(lambda x: tf.abs(x[0] - x[1]))([emb_a, emb_b])

out = layers.Dense(1, activation="sigmoid")(distance)

model = Model([input_a, input_b], out)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# Train
model.fit([x1, x2], y, epochs=15, batch_size=32, validation_split=0.2)

model.save("siamese_currency_model.h5")

print("Training complete!")
```
